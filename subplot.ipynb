````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
````

# ploting two subplots

````python
x1= np.arange(-10, 11, 1)
y1= np.power(x1, 2)

plt.figure(figsize= (7, 5))
plt.plot(x1, y1)
plt.grid()
plt.minorticks_on()
plt.title("y=x^2")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
````

````python
y2= np.power(x1, 3)

plt.figure(figsize= (7, 5))
plt.plot(x1, y2)
plt.grid()
plt.minorticks_on()
plt.title("y=x^3")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
````

````python
plt.figure(figsize= (7, 8))

# plot 1
plt.subplot(2, 1, 1)
x1= np.arange(-10, 11, 1)
y1= np.power(x1, 2)

plt.plot(x1, y1)
plt.grid()
plt.minorticks_on()
plt.title("y=x^2")
plt.xlabel("x")
plt.ylabel("y")

# plot 2
plt.subplot(2, 1, 2)
y2= np.power(x1, 3)

plt.plot(x1, y2)
plt.grid()
plt.minorticks_on()
plt.title("y=x^3")
plt.xlabel("x")
plt.ylabel("y")

plt.show()
````

# ploting side by side

````python
plt.figure(figsize= (12, 6))

# plot 1
plt.subplot(1, 2, 1)
x1= np.arange(-10, 11, 1)
y1= np.power(x1, 2)

plt.plot(x1, y1)
plt.grid()
plt.minorticks_on()
plt.title("y=x^2")
plt.xlabel("x")
plt.ylabel("y")

# plot 2
plt.subplot(1, 2, 2)
y2= np.power(x1, 3)

plt.plot(x1, y2)
plt.grid()
plt.minorticks_on()
plt.title("y=x^3")
plt.xlabel("x")
plt.ylabel("y")

plt.show()
````

# ploting subplots
```plt.subplots()``` method

````python
df = pd.read_csv('https://raw.githubusercontent.com/elyorbek8/data_visualization/refs/heads/main/merc.csv')
df.head()
````

````python
fig1, ax = plt.subplots(2, 2, figsize= (10, 12))
fig1.suptitle('Mercedes Benz plots')

sns.histplot(data= df, x= 'price', ax= ax[0, 0], hue= 'transmission', multiple= 'stack')
ax[0, 0].set_title('Price distribution')

sns.scatterplot(data= df, x= 'price', y= 'year', ax= ax[0, 1], hue= 'fuelType')
ax[0, 1].set_title('Price vs Year')

sns.histplot(data= df, x= 'mileage', ax= ax[1, 0], hue= 'transmission', multiple= 'stack')
ax[1, 0].set_title('Mileage distribution')

sns.scatterplot(data= df, x= 'mileage', y= 'price', ax= ax[1, 1], hue= 'fuelType')
ax[1, 1].set_title('Mileage vs Price')

plt.show()
````

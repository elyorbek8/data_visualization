```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/refs/heads/main/uzbekistan.csv', index_col = 0)
df
```

# Population distribution barplot

```python
plt.figure(figsize = (12, 6))
sns.barplot(x = df.index, y= df['Aholisi'], hue = df.index)

plt.title('Population Distribution in Uzbekistan')
plt.xlabel('Regions')
plt.ylabel('Population (millions)')
plt.grid()
plt.xticks(rotation = 90)
plt.show()
```

# Areas of regions in Uzbekistan

```python
plt.figure(figsize = (12, 6))
sns.barplot(x = df.index, y= df['Maydoni (kv.km)'], hue = df.index)

plt.title('Areas of regions in Uzbekistan')
plt.xlabel('Regions')
plt.ylabel('Areas (sq km)')
plt.grid()
plt.xticks(rotation = 90)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/elyorbek8/data_visualization/refs/heads/main/merc.csv')
df.head()
```

```python
df.info()
```

```python
sns.histplot(data= df, x= 'price')
plt.show()
```

```python
sns.displot(data= df, x= 'price', height= 6, aspect= 2)
plt.show()
```

# adding hue

```python
sns.histplot(data= df, x= 'price', hue= 'fuelType')
plt.show()
```

```python
sns.histplot(data= df, x= 'price', hue= 'transmission')
plt.show()
```

```python
sns.histplot(data= df, x= 'price', hue= 'fuelType', multiple= 'stack')
plt.show()
```

```python
sns.histplot(data= df, x= 'price', hue= 'transmission', multiple= 'stack')
plt.show()
```

```python
sns.displot(data= df, x= 'price', col= 'transmission', hue= 'fuelType', multiple= 'stack')
plt.show()
```

# KDE

```python
sns.displot(data= df, x= 'price', kind= 'kde', height= 5, aspect= 2)
plt.show()
```

```python
sns.displot(data= df, x= 'price', kde= True, height= 5, aspect= 3)
plt.show()
```

```python
sns.displot(data= df, x= 'price', kind= 'hist', height= 5, aspect= 2)
plt.show()
```

```python
sns.kdeplot(data= df, x= 'price', shade= True)
plt.show()
```

```python
sns.kdeplot(data= df, x= 'price', shade= True, hue= 'transmission')
plt.show()
```

# CDF plot

```python
sns.displot(data= df, x= 'price', kind= 'ecdf', height= 5, aspect= 2)
plt.grid()
plt.show()
```

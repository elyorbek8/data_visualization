# pylot line graph

```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

# ploting a simple line graph

```python
numbers = np.arange(12)
numbers
```

```python
plt.figure(figsize= (4,4))
plt.plot(numbers, color = 'g')
plt.show()
```

# ploting a sin(x) graph

```python
plt.figure(figsize = (10, 4))

x = np.arange(-2* np.pi, 2* np.pi, 0.1)
y = np.sin(x)
plt.plot(
    x, y,
    linewidth = 2,
    linestyle = '--',
    color = 'b',
    marker = 'o'
    )
plt.title('sin(x) function')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.grid()
plt.show()
```

# ploting both sin(x) and cos(x) in one graph

```python
x = np.arange(-2* np.pi, 2* np.pi, 0.1)
sin = np.sin(x)
cos = np.cos(x)

plt.figure(figsize= (10, 4))

# sin(x) graph
plt.plot(x, sin,
         color = 'r',
         linewidth = 2,
         linestyle = '--',
        #  label = 'sin(x)',
         marker = 'o',
         alpha = 0.8
         )
# cos(x) graph
plt.plot(x, cos,
         color = 'b',
         linewidth = 2,
         linestyle = '--',
        #  label = 'cos(x)',
         marker = '^',
         alpha = 0.8
         )

plt.title('sin(x) and cos(x) functions')
plt.xlabel('x')
plt.ylabel('functions')
plt.grid()
plt.legend(['sin(x)', 'cos(x)'], loc= 0)
plt.savefig('graph1.png')
plt.show()
```

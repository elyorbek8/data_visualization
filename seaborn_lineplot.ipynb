```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/refs/heads/main/museum_visitors.csv')
df
```

```python
df_long = df.melt(id_vars = 'Date', var_name = 'Museum', value_name = 'Visitors')
df_long
```

# seaborn lineplot

```python
df_long.info()
```

```python
plt.figure(figsize= (16, 6))
sns.lineplot(data = df_long, x = 'Date', y= 'Visitors', hue = 'Museum')
plt.xticks(rotation = 90)
plt.title('Museum visits over yaers')
plt.grid()
plt.show()
```

# matplotlib pyplot line plot

```python
df.columns
```

```python
plt.figure(figsize = (16, 6))

plt.plot(df['Date'], df['Avila Adobe'])
plt.plot(df['Date'], df['Firehouse Museum'])
plt.plot(df['Date'], df['Chinese American Museum'])
plt.plot(df['Date'], df['America Tropical Interpretive Center'])
plt.xlabel('Date')
plt.legend(['Avila Adobe', 'Firehouse Museum', 'Chinese American Museum',
       'America Tropical Interpretive Center'])
plt.xticks(rotation = 90)
plt.grid()
plt.show()
```

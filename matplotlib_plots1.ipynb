# matplotlib pyplot plot

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# I need to plot a line through the two points: (2, 3) and (7, 12)

x_points = np.array([2,7])
y_points = np.array([3, 12])

# plt.plot(x_points, y_points, marker= 'o', color= 'green', linestyle= 'dashed')
plt.plot(x_points, y_points, 'bo--')
plt.show()
```

```python
# drawing through multiple points
x = np.array([1, 3, 3.5, 5, 7, 9])
y = np.array([4, 12, 9, 17, 7, 18])

# plt.plot(x, y, '--Dc', markersize=15)
plt.plot(x, y, c= '#4CAF50', ms= 14, markeredgecolor= 'r', markerfacecolor= 'c', ls= 'solid', marker= 'o', linewidth= .9)
plt.show()
```

# ploting multiple lines

```python
y1 = np.array([1, 3, 3.5, 5, 7, 9])
y2 = np.array([4, 12, 9, 17, 7, 18])

plt.plot(y2,  's-b', label= 'line 1')
plt.plot(y1, 'o-g', label= 'line 2')
plt.legend()

font1 = {'family':'serif','color':'blue','size':15}
font2 = {'family':'serif','color':'darkred','size':15}
font3 = {'family':'sans-serif','color':'c','size':16}

plt.xlabel('x values', fontdict= font1)
plt.ylabel('y values', fontdict= font2)
plt.title('Testing lines', fontdict= font3, loc= 'left')
# plt.grid(axis='x')
# plt.grid(axis= 'y')
plt.grid(axis= 'both', c='orange', ls='-.', lw=0.2)

plt.show()
```

# Subplot

```python
# plot 2
x = np.array([1, 3, 3.5, 5, 7, 9])
y = np.array([4, 12, 9, 17, 7, 18])

plt.subplot(1, 2, 2)
plt.plot(x, y, c= '#4CAF50', ms= 14, markeredgecolor= 'r', markerfacecolor= 'c', ls= 'solid', marker= 'o', linewidth= .9)


# plot 1
y1 = np.array([1, 3, 3.5, 5, 7, 9])
y2 = np.array([4, 12, 9, 17, 7, 18])

plt.subplot(1, 2, 1)
plt.plot(y2,  's-b', label= 'line 1')
plt.plot(y1, 'o-g', label= 'line 2')
plt.legend()

font1 = {'family':'serif','color':'blue','size':15}
font2 = {'family':'serif','color':'darkred','size':15}
font3 = {'family':'sans-serif','color':'c','size':16}

plt.xlabel('x values', fontdict= font1)
plt.ylabel('y values', fontdict= font2)
plt.title('Testing lines', fontdict= font3, loc= 'left')
# plt.grid(axis='x')
# plt.grid(axis= 'y')
plt.grid(axis= 'both', c='orange', ls='-.', lw=0.2)

plt.show()
```

# Subplot: on top of each other

```python
# plot 1
x = np.array([1, 3, 3.5, 5, 7, 9])
y = np.array([4, 12, 9, 17, 7, 18])

plt.subplot(2, 1, 1)
plt.plot(x, y, c= '#4CAF50', ms= 14, markeredgecolor= 'r', markerfacecolor= 'c', ls= 'solid', marker= 'o', linewidth= .9)


# plot 2
y1 = np.array([1, 3, 3.5, 5, 7, 9])
y2 = np.array([4, 12, 9, 17, 7, 18])

plt.subplot(2, 1, 2)
plt.plot(y2,  's-b', label= 'line 1')
plt.plot(y1, 'o-g', label= 'line 2')
plt.legend()

font1 = {'family':'serif','color':'blue','size':15}
font2 = {'family':'serif','color':'darkred','size':15}
font3 = {'family':'sans-serif','color':'c','size':16}

plt.xlabel('x values', fontdict= font1)
plt.ylabel('y values', fontdict= font2)
# plt.title('Testing lines', fontdict= font3, loc= 'left')
# plt.grid(axis='x')
# plt.grid(axis= 'y')
plt.grid(axis= 'both', c='orange', ls='-.', lw=0.2)

plt.show()
```

# Setting particular colors for each dot

```python
x = np.array([1, 3, 5, 8, 2, 17, 12, 9, 4, 11, 12, 9, 6])
y = np.array([5, 12, 10, 16, 11, 29, 10, 17, 24, 18, 37, 15, 16])

colors = np.array(["blue","yellow","pink","black","orange","purple","beige","brown","gray","cyan","magenta", 'blue', 'yellow'])
plt.scatter(x, y, c= colors)
plt.show()
```

# Setting colors using color map: viridis

```python
x = np.array([1, 3, 5, 8, 2, 17, 12, 9, 4, 11, 12])
y = np.array([5, 12, 10, 16, 11, 29, 10, 17, 24, 18, 37])

colors = np.arange(0, 101, 10)
plt.scatter(x, y, c= colors, cmap= 'viridis')
plt.colorbar()

plt.show()
```

# Setting sizes for dots

```python
x = np.array([1, 3, 5, 8, 2, 17, 12, 9, 4, 11, 12])
y = np.array([5, 12, 10, 16, 11, 29, 10, 17, 24, 18, 37])

colors = np.arange(0, 101, 10)
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600])

plt.scatter(x, y, c= colors, cmap= 'viridis', s= sizes)

plt.show()
```

# Setting tranparency

```python
x = np.array([1, 3, 5, 8, 2, 17, 12, 9, 4, 11, 12])
y = np.array([5, 12, 10, 16, 11, 29, 10, 17, 24, 18, 37])

sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600])

plt.scatter(x, y, s= sizes, alpha= 0.4)

plt.show()
```

```python
# all in one
x = np.random.randint(100, size= (80))
y = np.random.randint(100, size= (80))

colors = np.random.randint(100, size= (80))
sizes = np.random.randint(300, size= (80))

plt.scatter(x, y, c= colors, cmap= 'nipy_spectral', s= sizes, alpha= 0.6)
plt.colorbar()

plt.show()
```

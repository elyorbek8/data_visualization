```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/elyorbek8/data_visualization/refs/heads/main/museum_visitors.csv', index_col = 0, usecols= [0, 4], parse_dates= [0])
df.head()
```

```python
type(df.index)
```

```python
df['month'] = [i.month for i in df.index]
df.head()
```

```python
df['year'] = [i.year for i in df.index]
df.head()
```

```python
df2 = df.groupby(['month', 'year']).mean()
df2
```

```python
df3 = df2.unstack(level = 0)
df3
```

```python
plt.figure(figsize = (10, 6))

hmap1 = sns.heatmap(df3, cmap='Blues', linewidth = 1, square = True)
plt.title('Visits to America Tropical Interpretive Center')
xticks_labels = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
plt.xticks(np.arange(12) + 0.5, xticks_labels, rotation = 45, ha= 'right')
hmap1.set_xlabel('Months')
hmap1.set_ylabel('Years')
plt.show()
```
